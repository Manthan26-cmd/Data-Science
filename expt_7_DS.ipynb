```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/water_potability.csv')
```

```python
data.head()
```

```python
from sklearn.feature_selection import chi2, f_classif, mutual_info_classif, VarianceThreshold
from sklearn.preprocessing import MinMaxScaler
```

```python
# Fill missing values with column mean
data_filled = data.fillna(data.mean(numeric_only=True))
```

```python
# Separate features and target
X = data_filled.drop(columns=["Potability"])
y = data_filled["Potability"]
```

```python
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
# 1. Correlation Coefficient (Pearson)
correlations = data_filled.corr()["Potability"].drop("Potability")
```

```python
# 2. Chi-Square Test
chi2_vals, _ = chi2(X_scaled, y)
```

```python
# 3. ANOVA F-test
f_vals, _ = f_classif(X_scaled, y)
```

```python
# Combine all results into one DataFrame
feature_scores = pd.DataFrame({
    "Correlation (Pearson)": correlations,
    "Chi2 Score": chi2_vals,
    "ANOVA F-score": f_vals
})
```

```python
feature_scores.sort_values(by="ANOVA F-score", ascending=False)
```
