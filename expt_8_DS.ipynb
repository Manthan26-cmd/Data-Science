```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/water_potability.csv')
```

```python
data.head()
```

```python
from sklearn.feature_selection import chi2, f_classif, mutual_info_classif, VarianceThreshold
from sklearn.preprocessing import MinMaxScaler
```

```python
# Fill missing values with column mean
data_filled = data.fillna(data.mean(numeric_only=True))
```

```python
# Separate features and target
X = data_filled.drop(columns=["Potability"])
y = data_filled["Potability"]
```

```python
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
# 1. Mutual Information
mi_vals = mutual_info_classif(X_scaled, y, random_state=42)
```

```python
# 5. Variance Threshold
selector = VarianceThreshold(threshold=0.01)
selector.fit(X_scaled)
variances = X_scaled.var()
```

```python
# Combine all results into one DataFrame
feature_scores = pd.DataFrame({
    "Mutual Info": mi_vals,
    "Variance": variances
})
```

```python
feature_scores.sort_values(by="Mutual Info", ascending=False)

```
